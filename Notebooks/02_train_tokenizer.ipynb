```python
from datasets import load_dataset
```

```python
dataset = load_dataset("roneneldan/TinyStories", split="train", streaming=True)
```

```python
texts = []
for i, item in enumerate(dataset):
    if i >= 100_000:
        break
    texts.append(item["text"])

print(f"Loaded {len(texts)} stories")
```

```python
subset_text = "\n".join(texts)
```

```python
from tokenizer import BPETokenizer
from config import GPTConfig
```

```python
tokenizer = BPETokenizer()
tokenizer.train(subset_text, vocab_size=GPTConfig.vocab_size)
```

```python
text = "Shahing"
ids = tokenizer.encode(text)
print(ids)
print(tokenizer.decode(ids))

```

```python
tokenizer.save("data/vocab.json", "data/merges.json")
```

```python
tokenizer2 = BPETokenizer()
tokenizer2.load("data/vocab.json", "data/merges.json")
assert tokenizer2.encode("hello world") == tokenizer.encode("hello world")
print("Save/load verified!")
```
